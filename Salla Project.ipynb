### Retrieve all records related to the agent named "Rania"

```python

%%sql
select * from Salla_Data Where Agent_Name='Rania';
```

### Display the number of calls offered for each day.

```python
%%sql
select date,sum(calls_offered) as 'Call Ofeered' from Salla_Data
group by date order by date desc;
```

### List all unique agent names.

```python
%%sql
select distinct agent_name from salla_data;
```

### Calculate the average ASA (Average Speed of Answer)

```python
%%sql
select AVG(asa) as 'average ASA' from salla_data;
```

### Sort the data by the number of abandoned calls in descending order.

```python
%%sql
select * from salla_data order by calls_abandon desc;
```

### Calculate the total forecasted calls and total offered calls for the month of February.

```python
%%sql
select sum(forecasted_calls)as'total forecasted calls',
sum(calls_offered)as'total offered calls' from salla_data where month='22-Feb';
```

### For each agent, calculate the call handling rate as a percentage of handled calls to offered calls.

```python
%%sql
SELECT agent_name,(CAST(SUM(calls_handled) AS FLOAT) / CAST(SUM(calls_offered) AS FLOAT) * 100)AS 'the call handling rate'
FROM salla_data
GROUP BY agent_name;
```

### classify each day's ASA as:"Excellent" if ASA < 5"Good" if ASA is between 5 and 10 "Poor" if ASA > 10

```python
%%sql
select date , case
                when asa < 5 then 'Excellent' 
				when asa between 5 and 10 then 'Good' 
				when asa >10 then 'Poor' 
				end as 'status'
				from salla_data;
```

### Show the records where the number of abandoned calls is above the average number of abandoned calls.

```python
%%sql
select * from salla_data where calls_abandon > (select avg(calls_abandon) from salla_data);
```
